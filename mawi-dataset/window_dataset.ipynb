```python
import csv
import pandas as pd

output = open('edit.csv', 'w')
writer = csv.writer(output)

step = 5.0
actual_step = 5.0

with open('20190105_clean.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    bytes = []
    results = []
    first_line = True

    for row in csv_reader:
        if(first_line == True): 
            first_line = False
            continue

        time = row[0]
        b = row[1]
        
        if float(time) > actual_step:
            results.append(sum(bytes))
            actual_step += step
            bytes.clear()
        
        bytes.append(float(b))
            
csv_file.close()
output.close()

# print(results)

# print(len(results))

from matplotlib import pyplot as plt
plt.plot(results)
plt.show()

with open("clean_dataset.csv", "w") as file:
    file.write("Step,B\n")
    for i, r in enumerate(results):
        file.write("{},{}\n".format(i*step,r))

# df = pd.read_csv('edit.csv')
# df.to_csv('edit.csv', index=False)
```
